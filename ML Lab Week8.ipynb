```python
'''Take two 3D coordinates from the user. Find out the distance between these points
using following distance functions.
Distance functions
Euclidean
Manhattan
Minkowski
x and y are vectors representing the coordinates of two points.
Here, k=3'''

import numpy as np

def euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y) ** 2))

def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

def minkowski_distance(x, y, q):
    return np.power(np.sum(np.power(np.abs(x - y), q)), 1/q)
def get_coordinates():
    coordinates = []
    for i in range(2):
        point = []
        for j in range(3):
            coord = float(input(f"Enter coordinate {j+1} of point {i+1}: "))
            point.append(coord)
        coordinates.append(point)
    return np.array(coordinates)

print("Enter the coordinates for the two points:")
x = get_coordinates()
y = get_coordinates()
euclidean_dist = euclidean_distance(x, y)
manhattan_dist = manhattan_distance(x, y)
q = float(input("Enter the value of q for Minkowski distance: "))
minkowski_dist = minkowski_distance(x, y, q)
print(f"Euclidean distance: {euclidean_dist}")
print(f"Manhattan distance: {manhattan_dist}")
print(f"Minkowski distance (q={q}): {minkowski_dist}")
```

```python
'''Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for
all the data samples in the feature space. Distance metric is a 2D array, where the
(i,j)th entry represents the distance between the ith and jth sample points in the feature
space.'''

from sklearn.datasets import load_iris
from scipy.spatial.distance import euclidean
import numpy as np

iris = load_iris()
data = iris.data

def euclidean_distance_matrix(data):
    num_samples = len(data)
    distance_matrix = np.zeros((num_samples, num_samples))
    for i in range(num_samples):
        for j in range(num_samples):
            distance_matrix[i, j] = euclidean(data[i], data[j])
    return distance_matrix
distance_matrix = euclidean_distance_matrix(data)
print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
'''Import the Iris dataset. Prepare a dataset considering samples belong to any two
output classes. Draw the scatter plot for all the samples in the new dataset considering
any two input attributes. Examine the scatter plot to find the equation of a line that
can separate sample of two classes.'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data
target = iris.target
class1 = 0  # Class 1
class2 = 1  # Class 2
class1_indices = np.where(target == class1)[0]
class2_indices = np.where(target == class2)[0]
class1_data = data[class1_indices]
class2_data = data[class2_indices]
feature1_index = 0  
feature2_index = 1  
plt.figure(figsize=(8, 6))
plt.scatter(class1_data[:, feature1_index], class1_data[:, feature2_index], color='blue', label=f'Class {class1}')
plt.scatter(class2_data[:, feature1_index], class2_data[:, feature2_index], color='red', label=f'Class {class2}')
plt.xlabel(f'Feature {feature1_index + 1}')
plt.ylabel(f'Feature {feature2_index + 1}')
plt.title(f'Scatter Plot of Classes {class1} and {class2} using Features {feature1_index + 1} and {feature2_index + 1}')
plt.legend()
plt.grid(True)
plt.show()
```
